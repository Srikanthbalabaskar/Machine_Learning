```python
import pandas as pd
import numpy as np
```

```python
loan=pd.read_csv('loan_data.csv')
```

```python
loan.head()
```

```python
loan.shape
```

```python
loan.describe()
```

```python
loan.dtypes
```

```python
loan.isna().sum()
```

```python
encodedDF=pd.get_dummies(loan['purpose'])
```

```python
encodedDF.head()
```

```python
X=pd.concat([encodedDF,loan],axis=1)
```

```python
X.head()
```

```python
X.drop(['purpose'],axis=1,inplace=True)
```

```python
X.head()
```

```python
X.shape
```

```python
y=X.pop('not.fully.paid')
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(train,y_train)
y_pred=lr.predict(test)
print("Accuracy Score Baseline:", accuracy_score(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=4)
dt.fit(train,y_train)
y_predDecsion=dt.predict(test)
print(accuracy_score(y_test,y_predDecsion))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(train,y_train)
pred_rf=rf.predict(test)
print(accuracy_score(y_test,pred_rf))
```
